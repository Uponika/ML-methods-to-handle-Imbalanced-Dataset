```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("seismic-bumps.csv")
data.head()
```

```python
data.shape
data['class'].value_counts()
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['seismic']= label_encoder.fit_transform(data['seismic']) 
  
data['seismic'].unique() 
```

```python

data['seismoacoustic']= label_encoder.fit_transform(data['seismoacoustic']) 
data['ghazard']= label_encoder.fit_transform(data['ghazard']) 
data['shift']= label_encoder.fit_transform(data['shift']) 

```

```python
data.head()
```

```python
X = data.drop(['id', 'class'], axis=1)
y= data['class']
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import cross_val_score
data_Check = LogisticRegressionCV(class_weight='balanced', random_state = 20)
scores = cross_val_score(data_Check, X, y, cv=5)
print(scores)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=3)
for train, test in skf.split(X, y):
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
```
